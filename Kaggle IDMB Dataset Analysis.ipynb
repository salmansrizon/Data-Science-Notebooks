```python
import pandas as pd
#read the dataset
excel_file = 'movies.xls'
movies = pd.read_excel(excel_file)
#Preview the dataset
movies.head()

```

```python
#to read multiple sheet in the excel file
movies_sheet1 = pd.read_excel(excel_file, sheetname=0, index_col=1)
#Read the sheet 2 of the file
movies_sheet2 = pd.read_excel(excel_file, sheetname=1, index_col=1)
#Read the sheet 3 of the file 
movies_sheet3 = pd.read_excel(excel_file, sheetname=0, index_col=1)
#preview the data 
movies_sheet2.head()
```

```python
#Concating the each sheet of the file
movies = pd.concat([movies_sheet1,movies_sheet2,movies_sheet3])

#Total Number of row and colums
movies.shape
```

```python
#View the bottom row of the data table
movies_sheet2.tail()
```

```python
#Sorted Movies by Gross Earnings
sorted_by_gross = movies.sort_values(["Gross Earnings"], ascending=False)
#Adding Matplot lib for plotting the data in graph
import matplotlib.pyplot as plt
sorted_by_gross['Gross Earnings'].head(20).plot(kind="barh")
plt.show()
```

```python
#Sorting According to movie Budget
sorted_by_budget = movies.sort_values(["Budget"], ascending=False)
sorted_by_budget.head(10)
```

```python
sorted_by_budget['Budget'].head(10).plot(kind="barh")
plt.show()
```

```python
#Sorting according to IMDB score
movies['IMDB Score'].plot(kind="hist")
plt.show()
```

```python
#Appliying formulas on colums
movies["Net Earnings"] = movies["Gross Earnings"] - movies["Budget"]

sorted_by_net_income = movies[['Net Earnings']].sort_values(['Net Earnings'], ascending=[False])
sorted_by_net_income.head(10)['Net Earnings'].plot.barh()
```
